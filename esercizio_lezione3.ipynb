Sistema di Gestione Voti Studenti Realizza un programma Python che permetta di:

Inserire una serie di voti (numeri interi) Analizzare i voti inseriti Mostrare un riepilogo finale

Inserimento dati

Chiedere all’utente di inserire voti uno alla volta da terminale
Salvare i voti in una lista
L’inserimento termina quando l’utente digita: fine
Durante l’inserimento: Il voto deve essere:

numerico
compreso tra 0 e 30 In caso di errore:
mostrare un messaggio
NON aggiungere il valore alla lista
Elaborazione dati Utilizzando un ciclo for, calcolare:

Somma totale dei voti
Media dei voti
Numero di voti sufficienti (≥ 18)
Numero di voti insufficienti (< 18)
Voto massimo
Voto minimo
Usare if-elif-else per classificare la media:

Media ≥ 27 → "Eccellente"
Media ≥ 21 → "Buono"
Media ≥ 18 → "Sufficiente"
Media < 18 → "Insufficiente"
Output richiesto

Stampare:

Lista voti inseriti
Media
Numero sufficienti / insufficienti
Voto massimo e minimo
Classificazione finale

```python
voti = []

while True:
    valore = input("Inserisci un voto e digita 'fine' per terminare: ")
    #controlla se il ciclo deve terminare
    if valore.lower() == 'fine':
        print("Fine inserimento...")
        break
    
    if valore.isdigit():  #controlla se è un numero
        voto = int(valore)
        #compreso tra 0 e 30
        if voto > 0 and voto <= 30:
            voti.append(voto) #serve per aggiungere un elemento alla lista
        else:
            print("Devi inserire un valore compreso tra 0 e 30")
    else:
        print("Devi inserire un valore numerico, riprova!")
        
print("Inserimento voti terminato.")

#analizzare i voti inseriti
if len(voti) == 0:
    print("Nessun voto inserito...")
else:
    #somma totale dei voti
    somma = 0 #somma totale dei voti
    sufficienti = 0 #numero di voti sufficienti (>=18)
    insufficienti = 0 #numero di voti insufficienti (<18)
    massimo = voti[0] #voto massimo
    minimo = voti[0] #voto minimo
    
    [12, 18, 24, 22, 30]
    for v in voti:
        somma += v #somma dei voti
        
        if v >= 18:
            sufficienti += 1  #conto voti sufficienti
        else:
            insufficienti += 1  #conto voti insufficienti
            
        if v > massimo:
            massimo = v  #calcolo il voto massimo
        
        if v < minimo:
            minimo = v  #calcolo il voto
    
    media = somma / len(voti) #calcola la media dei voti
    
if media >= 27:
    classificazione = "Eccellente"
elif media >= 21:
    classificazione = "Buono"
elif media >=18:
    classificazione = "Sufficiente"
else:
    classificazione = "Insufficiente"
    
#output richiesto
#stampare
print(voti)
print("Media dei voti:", media)
print("Numero di voti sufficienti:", sufficienti)
print("Numero di voti insufficienti:", insufficienti)
print("Somma totale dei voti:", somma)
print("Classificazione:", classificazione)
#analizzare i voti inseriti
#mostrare un riepilogo finale





```
