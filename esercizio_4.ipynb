# Un piccolo negozio di libri desidera gestire il proprio catalogo 
# in maniera semplice. 
# Ogni libro è rappresentato da una tupla contenente le seguenti 
# informazioni: 
# (titolo: str, autore: str, anno_pubblicazione: int, prezzo: float)
# Scrivi uno script Python che soddisfi i seguenti requisiti:
# - Crea una lista di almeno 5 libri, ciascuno rappresentato 
#   come una tupla nel formato sopra indicato.
# - Stampa tutti i libri presenti nel catalogo, uno per riga, 
#   formattando le informazioni in modo leggibile 
#   (es. “Titolo: ..., Autore: ..., Anno: ..., Prezzo: ...”).
# - Chiedi all’utente un anno, quindi stampa tutti i libri 
#   pubblicati dopo quell’anno.
# - Chiedi all’utente il nome di un autore e mostra 
#   tutti i libri scritti da quell’autore.
# - Calcola e stampa il prezzo medio dei libri presenti nel catalogo.
# - Trova e stampa il libro più costoso nel catalogo.
# - Poiché le tuple sono immutabili, spiega in un commento come 
#   potresti aggiornare il prezzo di un libro 

```python
catalogo = [
    ("The Great Gatsby", "F. Scott Fitzgerald", 1925, 12.50),
    ("To Kill a Mockingbird", "Harper Lee", 1960, 15.00),
    ("1984", "George Orwell", 1949, 14.00),
    ("Pride and Prejudice", "Jane Austen", 1813, 13.50),
    ("The Catcher in the Rye", "J.D. Salinger", 1951, 16.00)
]
print("Catalogo dei libri:")
for libro in catalogo:
    # titolo, autore, anno, prezzo = libro
    # print(f"Titolo:  {titolo}")
    # print(f"Autore: {autore}")
    # print(f"Anno: {anno}")
    # print(f"Prezzo: ${prezzo:.2f}")
    # print("-------------------------------")
    # print("-------------------------------")
    # print("Ricerca libri per anno di pubblicazione")        CTRL + ù PER COMMENTARE
    
    (titolo, autore, anno, prezzo) = libro
    print(f"Titolo:  {titolo}, Autore: {autore}, Anno: {anno}, Prezzo: ${prezzo:.2f}")
print("---------------------------------------------------------------")
anno_pubblicazione = int(input("Inserisci l'anno di pubblicazione per la ricerca: "))
libri_trovati = False
print(f"Libri pubblicati dopo il {anno_pubblicazione}:")
for libro in catalogo:
    (titolo, autore, anno_pubblicazione, prezzo) = libro
    if anno_pubblicazione > anno:
        print(f"  -Titolo:  {titolo}, Autore: {autore}, Anno: {anno_pubblicazione}, Prezzo: ${prezzo:.2f}")
        libri_trovati = True
if not libri_trovati:
    print(f"  ---Non ci sono libri pubblicati dopo il {anno_pubblicazione}---  ")

print("---------------------------------------------------------------")


# Chiedi all’utente il nome di un autore e mostra 
# tutti i libri scritti da quell’autore.
nome_autore = input("Inserisci il nome dell'autore per la ricerca: ").strip().capitalize()
print(f"Libri pubblicati da {nome_autore}:")
libri_trovati = False

for libro in catalogo:
    (titolo, autore, anno_pubblicazione, prezzo) = libro
    if autore.lower() == nome_autore.lower():
        print(f"  -Titolo:  {titolo}, Autore: {autore}, Anno: {anno_pubblicazione}, Prezzo: ${prezzo:.2f}")
        libri_trovati = True

if not libri_trovati:
    print(f"  ---Non ci sono libri pubblicati da {nome_autore}---  ")
    
print("---------------------------------------------------------------")

sommaPrezzoLibri = 0
for libro in catalogo:
    sommaPrezzoLibri += libro[3]
    
prezzoMedio = sommaPrezzoLibri / len(catalogo)
print(f"Il prezzo medio dei libri nel catalogo è: ${prezzoMedio:.2f}")

print("---------------------------------------------------------------")

vechioLibro = catalogo[0]

nuovoLibro[3] = 10.00
catalogo[0] = nuovoLibro
print("Catalogo aggiornato:")
```
