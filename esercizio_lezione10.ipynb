# List Comprehension
# Crea una lista contenente solo i nomi degli studenti sufficienti (voto ≥ 18)
# Crea una lista con i voti aumentati di 1 punto, ma solo per chi ha voto ≥ 18
# Crea una lista di stringhe nel formato: "Nome - Promosso/Bocciato"

# Set Comprehension
# Crea un set con tutti i voti unici
# Crea un set contenente solo i voti insufficienti

# Dictionary Comprehension
# Crea un dizionario: {nome: voto}
# Crea un dizionario contenente solo gli studenti promossi
# Crea un dizionario: {nome: "Eccellente" | "Buono" | "Sufficiente" | "Insufficiente"}
# usando le seguenti regole:
# ≥ 27 → Eccellente
# ≥ 21 → Buono
# ≥ 18 → Sufficiente
# < 18 → Insufficiente

```python
from pprint import pprint

studenti = [
    {"Nome": "Luca", "voto": 28},
    {"Nome": "Anna", "voto": 17},
    {"Nome": "Marco", "voto": 30},
    {"Nome": "Giulia", "voto": 22},
    {"Nome": "Paolo", "voto": 18},
    {"Nome": "Sara", "voto": 15},
    {"Nome": "Mario", "voto": 22}
]

# LIST COMPREHENSION
sufficienti = [s["Nome"] for s in studenti if s["voto"] >= 18]

voti_aumentati = [s["voto"] + 1 for s in studenti if s["voto"] >= 18]

esiti = [
    f"{s['Nome']}: {'Promosso' if s['voto'] >= 18 else 'Bocciato'}"
    for s in studenti
]

# SET COMPREHENSION
voti_unici = {s["voto"] for s in studenti}

voti_insufficienti = {s["voto"] for s in studenti if s["voto"] < 18}

# DICTIONARY COMPREHENSION
dizionario_voti = {s["Nome"]: s["voto"] for s in studenti}

promossi = {s["Nome"]: s["voto"] for s in studenti if s["voto"] >= 18}

valutazione = {
    s["Nome"]: (
        "Eccellente" if s["voto"] >= 27 else
        "Buono" if s["voto"] >= 21 else
        "Sufficiente" if s["voto"] >= 18 else
        "Insufficiente"
    )
    for s in studenti
}

# STAMPA RISULTATI
print("==========Lista Studenti==========")
pprint(studenti, sort_dicts=False)
print("========================================================================================")
print("============Risultati============")
print("Studenti con voto sufficiente:", sufficienti)
print("Voti aumentati:", voti_aumentati)
print("Esiti:", esiti)
print("Voti unici:", voti_unici)
print("Voti insufficienti:", voti_insufficienti)
print("Studenti promossi:", promossi)
print("Valutazione:", valutazione)
```
