Sistema di Calcolo Sconto Cliente Realizza un programma Python che simuli un sistema di calcolo sconto per un negozio. Il programma deve:

Richiedere all’utente: nome cliente importo totale della spesa età del cliente

Applica uno sconto in base alle seguenti condizioni: Se l’importo è: ≥ 100€ → sconto 10% ≥ 50€ e < 100€ → sconto 5% < 50€ → nessuno sconto Sconto aggiuntivo: Se età < 18 → +5% Se età > 65 → +7%

Gli sconti sono cumulativi (somma delle percentuali)

Calcolare: sconto totale (%) valore sconto (€) prezzo finale

Stampare un riepilogo con:

Nome cliente
Importo iniziale
Percentuale sconto totale
Importo sconto
Prezzo finale

```python
nome = input("Inserisci il nome: ").strip()
cognome = input("Inserisci il cognome: ").strip()
importoTOT = float(input("Inserisci l'importo totale della spesa: "))
eta = int(input("Inserisci l'età: "))
if eta < 0:
    print("ERRORE: inserisci un'età valida!")
    exit()
    
sconto = 0 

if importoTOT >= 100:
    sconto += 0.10
elif importoTOT >= 50:
    sconto += 0.05

if eta < 18:
    sconto += 0.05
elif eta > 65:
    sconto += 0.07

valore_sconto = importoTOT * sconto
prezzo_finale = importoTOT - valore_sconto

print("\n--Ecco un riepilogo del tuo acquisto--")
print(f"Cliente: {nome} {cognome}")
print(f"Importo iniziale: {importoTOT:.2f} €")
print(f"Sconto totale: {sconto * 100:.1f}%")
print(f"Valore sconto: {valore_sconto:.2f} €")
print(f"Prezzo finale: {prezzo_finale:.2f} €")
```

Sistema di Accesso con Tentativi Limitati Realizza un programma Python che simuli un sistema di login da terminale.

Il programma deve: Richiedere all’utente: username e password Confrontare i dati inseriti con credenziali predefinite: username_correto = "admin" password_corretta = "1234"

L’utente ha massimo 3 tentativi Ad ogni errore:

mostrare un messaggio di errore
indicare i tentativi rimanenti Se le credenziali sono corrette:
stampare messaggio di accesso consentito
terminare il ciclo Se i tentativi finiscono:
bloccare l’accesso
stampare messaggio finale
Controlli obbligatori Username vuoto → messaggio di errore Password vuota → messaggio di errore

Output atteso (esempio) Tentativo 1/3 Username: admin Password: 1111 Credenziali errate. Tentativi rimasti: 2

Tentativo 2/3 Username: admin Password: 1234 Accesso consentito!

```python
username_corretto = "admin"
password_corretta = "1234"
max_tentativi = 3
tentativo = 0
tentativi_rimasti = max_tentativi - tentativo
username = input("\nInserisci il tuo username: ").strip()
password = input("Inserisci la tua password: ").strip()

while username != "admin":
    print("Username sbagliato, riprova!")
    tentativo += 1
    username = input("Inserisci il tuo username: ").strip()
while password != "1234":
    print("Password sbagliata, riprova!")
    tentativo += 1
    password = input("Inserisci la tua password: ").strip()
    print("Accesso riuscito! Benvenuto, admin.")
if tentativi_rimasti > 0:
    print(f"Tentativi rimasti: {tentativi_rimasti}\n")
else:
    print("Accesso bloccato. Hai esaurito i tentativi.")
    
    tentativo += 1
```
