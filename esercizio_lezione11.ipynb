Sistema di gestione di un conto bancario (uso di classi, incapsulamento e metodi dunder) Progettare una classe ContoBancario che rappresenti un conto corrente con le seguenti caratteristiche: -> Attributi (incapsulati): __titolare (stringa) __saldo (float) __movimenti (lista di stringhe che descrivono le operazioni)

-> Metodi pubblici: deposita(importo) Aggiunge denaro al saldo Registra il movimento Deve validare che l’importo sia positivo preleva(importo) Sottrae denaro dal saldo Impedisce saldo negativo Registra il movimento get_saldo() Restituisce il saldo attuale get_movimenti() Restituisce la lista dei movimenti

-> Requisiti sui metodi dunder: init(self, titolare, saldo_iniziale) Inizializza il conto Se il saldo iniziale è negativo → impostare a 0 str(self) Restituisce una stringa leggibile: Es: "Conto di Mario Rossi - Saldo: 1500€" repr(self) Restituisce una rappresentazione tecnica utile al debug eq(self, altro) Due conti sono uguali se: Hanno lo stesso titolare Hanno lo stesso saldo len(self) Restituisce il numero di movimenti effettuati

Output atteso (indicativo):

Saldo aggiornato correttamente
Movimenti registrati
Confronto tra oggetti funzionante
Rappresentazione leggibile e tecnica separate

```python
class ContoBancario:
    def __init__(self, titolare, saldo_iniziale):
        self.__titolare = titolare
        if saldo_iniziale < 0:
            self.__saldo = 0.0
        else:
            self.__saldo = float(saldo_iniziale)
            
        self.__movimenti = []
        self.__movimenti.append(f'Apertura conto con: {self.__saldo}£')
        
    def deposita(self, importo):
        if importo > 0:
            self.__saldo += importo
            self.__movimenti.append(f'Deposito: +{importo}£')
            print('Deposito di {importo}£ effettuato con successo.')
        else:
            print("Errore: l'importo da depositare deve essere positivo")
            
    def preleva(self, importo):
        if importo <= 0:
            print("Errore: l'importo da prelevare deve essere positivo")
        elif importo > self.__saldo:
            print("Errore: saldo insufficiente per l'operazione.")
        else:
            self.__saldo -= importo
            self.__movimenti.append(f"Prelievo: -{importo}£")
            print(f"Prelievo di {importo}£ effettuato con successo")
            
    def get_saldo(self):
        return self.__saldo
    
    def get_movimenti(self):
        return self.__movimenti
    
    def __str__(self):
        return f"Conto di {self.__titolare} - Saldo: {self.__saldo}£"
    
    def __repr__(self):
        return "Conto Bancario (titolare = {self.__titolare}", "saldo = {self.__saldo}, movimenti = {len(self.__movimenti)})"
    def __eq__(self, other):
        if not isinstance(other, ContoBancario):
            return False
        return self.__titolare == other.__titolare and self.__saldo == other.__saldo
    
    def __len__(self):
        return len(self.__movimenti)

# 1. Inizializzazione
conto1 = ContoBancario("Mario Rossi", 1500)
conto2 = ContoBancario("Francesca Cuccuru", 2000)
conto3 = ContoBancario("Michael Olise", -100)

# 2. Operazioni
conto1.deposita(500)
conto1.preleva(200)
conto1.preleva(1600)
# 3. Output saldo e movimenti
print(f"Saldo attuale: {conto1.get_saldo()}£")
print(f"Lista movimenti: {conto1.get_movimenti()}")
print(f"Numero totale movimenti: {len(conto1)}")

# 4. Verifica metodi Dunder
print("\n--- Verifica Metodi Speciali ---")
print(f"Metodo __str__: {conto1}")

# 5. Confronto tra oggetti
print("\n--- Confronto conti ---")
print(f"Conto1 è uguale a Conto2? {conto1 == conto2}") # False perché conto1 ha subito variazioni
conto4 = ContoBancario("Mario Rossi", 1800)
print(f"Conto1 è uguale a Conto4 (dopo operazioni)? {conto1 == conto4}") # True se il saldo finale coincide
```
