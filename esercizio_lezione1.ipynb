# Sistema Registrazione Utente
# Stai sviluppando un semplice programma da terminale 
# per registrare un utente e mostrare un riepilogo formattato.

# Il programma deve:
# Chiedere all’utente: nome, cognome, età
#Eseguire le seguenti operazioni:
# - Convertire l’età da stringa a intero
# - Calcolare la lunghezza del nome e cognome
# Trasformare:
# - nome → maiuscolo
# - cognome → minuscolo
# Stampare un riepilogo con:
# - nome completo formattato
# - età
# - lunghezza nome e cognome
# - tipo della variabile età

# Usare obbligatoriamente:
# - input()
# - print()
# - type()
# - len()
# Usare almeno 2 metodi delle stringhe:
# .upper()
# .lower()
# .strip()
# Inserire controlli if per:
# età negativa
# nome vuoto

```python
nome = input("Inserisci il tuo nome: ").strip()
if len(nome) ==0:
    print("Devi inserire un nome valido, capito???")
    

cognome = input("Inserisci il tuo cognome: ").strip()   
if len(cognome) ==0:
    print("Devi inserire un cognome valido, hai capito???")


eta = int(input("Inserisci la tua eta: ").strip())
if eta <0:
    print("Devi inserire un eta valido")
    

print(len(nome), len(cognome))
# print("Ciao " + nome.upper() + " " + cognome.lower() + ", hai " + str(eta) + " anni.")
# print("Nome completo: ", nome.upper(), cognome.lower())
# print("Eta: ", eta)
# print("Lunghezza nome: ", len(nome))
# print("Lunghezza cognome: ", len(cognome))
# print("Tipo della variabilita eta: ", type(eta))

print(f"Nome completo: {nome.upper()} {cognome.lower()}")
print(f"Eta: {eta}")
print(f"Lunghezza nome: {len(nome)}")
print(f"Lunghezza cognome: {len(cognome)}")
print(f"Tipo della variabilita eta: {type(eta)}")
```
